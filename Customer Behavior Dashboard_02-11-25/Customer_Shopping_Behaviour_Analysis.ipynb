### Filling the NA values from the Review Rating column by each Categories median values.

```python
a = df[["Category", "Review Rating"]]
a
```

```python
b = a[a["Review Rating"].isnull()]
b
```

```python
b.groupby("Category")["Review Rating"].size()
```

```python
df["Review Rating"] = df.groupby("Category")["Review Rating"].transform(lambda x : x.fillna(x.median()))
```

```python
df.isnull().sum()
```

### Snake Casing the Column Names for better readability.

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ", "_")
```

```python
df = df.rename(columns = {'purchase_amount_(usd)' : 'purchase_amount'})
```

```python
df.columns
```

### Creating a column Age Group

```python
labels = ["Young Adult", "Adult", "Middle_aged", "Senior"]
df["age_group"] = pd.qcut(df["age"], q = 4, labels = labels)
```

```python
df[["age","age_group"]].head(15)
```

```python
df.head()
```

### Creating a Column purchase_frequency_days

```python
df["frequency_of_purchases"].value_counts()
```

```python
frequency_mapping = {
    "Every 3 Months" : 90,
    "Annually" : 365,
    "Quarterly" : 90,
    "Monthly" : 30,
    "Bi-Weekly" : 14,
    "Fortnightly" : 14,
    "Weekly" : 7
}

df["purchase_frequency_days"] = df["frequency_of_purchases"].map(frequency_mapping)
```

```python
df[["frequency_of_purchases", "purchase_frequency_days"]]
```

### Checking discount_applied and promo_code_used columns for dropping one of them

```python
(df["promo_code_used"] == df["discount_applied"]).all()
```

```python
df.drop(columns = {"promo_code_used"}, axis = 1, inplace = True)
```
